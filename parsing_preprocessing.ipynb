```python
import re
from collections import defaultdict

def extract_all_roles(text):
    """
    Парсить усіх уповноважених осіб з їх ролями і датами.
    Створює словник: {роль: {"name": ..., "date": ...}}
    """
    if not isinstance(text, str):
        return {}

    # Очищення
    t = re.sub(r"\(Відомості відсутні\)|\(без обмежень\)", "", text, flags=re.I)
    t = re.sub(r"\s+", " ", t).strip()

    roles_list = [
        "керівник",
        "підписант",
        "ліквідатор",
        "виконуючий обов'язки",
        "голова комісії з припинення",
        "бухгалтер"
    ]
    roles_regex = "|".join(map(re.escape, roles_list))

    pattern = re.compile(
        rf"([А-ЯІЇЄҐA-Z'][А-ЯІЇЄҐA-Z' \-]+?)\s*(\d{{2}}\.\d{{2}}\.\d{{4}})?\s*,?\s*({roles_regex})",
        flags=re.I
    )

    found = defaultdict(dict)
    for name, date, role in pattern.findall(t):
        role = role.lower().strip()
        found[role]["name"] = name.strip(" ,;")
        if date:
            found[role]["date"] = date

    if not found and re.search(r"керівник", t, flags=re.I):
        m = re.search(r"([А-ЯІЇЄҐ][А-ЯІЇЄҐA-Z'\- ]+)\s*керівник", t)
        if m:
            found["керівник"]["name"] = m.group(1).strip()

    return found
```

```python
roles_expanded = df["Уповноважені особи"].apply(extract_all_roles)

for i, role_data in enumerate(roles_expanded):
    for role, info in role_data.items():
        name_col = role.capitalize()
        date_col = f"Дата_{role.capitalize()}"
        df.loc[i, name_col] = info.get("name")
        df.loc[i, date_col] = info.get("date")
```

### ВИДИ ДІЯЛЬНОСТІ

```python
import re
import numpy as np
import pandas as pd

def clean_activity_text(text):
    if not isinstance(text, str):
        return np.nan
    
    text = re.sub(r"Всього за цим КВЕД:.*?(ФОП|Фізичних осіб-підприємців)", "", text, flags=re.I)
    
    text = re.sub(r"Всі Види Діяльності.*", "", text, flags=re.I)
    text = re.sub(r"Приховати.*", "", text, flags=re.I)
    
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
def parse_activity_block(text):
    """Виділяє основний КВЕД і до трьох додаткових."""
    if not isinstance(text, str) or text.strip() == "":
        return pd.Series([np.nan]*8)

    text = clean_activity_text(text)
    
    m_main = re.search(r"Основний:\s*([\d.]+)\s+(.+?)(?=Інші:|$)", text, flags=re.I)
    main_code = m_main.group(1).strip() if m_main else np.nan
    main_name = m_main.group(2).strip() if m_main else np.nan
    
    others_block = re.search(r"Інші:\s*(.+)", text, flags=re.I)
    others = others_block.group(1).strip() if others_block else ""
    
    other_matches = re.findall(r"(\d{2}\.\d{2})\s+([^0-9]+?)(?=(?:\d{2}\.\d{2}|$))", others)
    
    others_limited = other_matches[:3]
    
    while len(others_limited) < 3:
        others_limited.append((np.nan, np.nan))
    
    result = [main_code, main_name]
    for code, name in others_limited:
        result.extend([code, name.strip() if isinstance(name, str) else np.nan])
    
    return pd.Series(result)
```

```python
df[[
    "KVED_main", "Activity_main",
    "KVED_2_CODE", "KVED_2_NAME",
    "KVED_3_CODE", "KVED_3_NAME",
    "KVED_4_CODE", "KVED_4_NAME"
]] = df["Види діяльності"].apply(parse_activity_block)
```

## Контактна інформація

```python
df["Контактна інформація"]
```

```python
def parse_contact_info(text):
    if not isinstance(text, str) or text.strip() == "":
        return pd.Series([np.nan, np.nan, np.nan])
    
    text = re.sub(
        r"В цей день ми додатково зробили перевірку.*",
        "",
        text,
        flags=re.I | re.S
    )
    
    text = re.sub(r"Місцезнаходження юридичної особи[:\-]?\s*", "", text, flags=re.I)
    phone_match = re.search(r"(\+?\d[\d\s\-()]{7,})", text)
    phone = phone_match.group(1).strip() if phone_match else np.nan
    
    email_match = re.search(r"[\w\.-]+@[\w\.-]+\.\w+", text)
    email = email_match.group(0).strip() if email_match else np.nan
    address = re.split(r"Телефон:|E-mail:|Email:", text, maxsplit=1)[0]
    address = re.sub(r"\s+", " ", address).strip(" ,;")
    
    return pd.Series([address, phone, email])

df[["Address", "Phone", "Email"]] = df["Контактна інформація"].apply(parse_contact_info)
df
```

## Перелік засновників

```python
df["Перелік засновників/учасників юридичної особи"]
```

```python
def parse_multiple_founders(text, max_founders=3):

    if not isinstance(text, str) or text.strip() == "":
        return pd.Series([np.nan] * (max_founders * 4))

    t = re.sub(r"\s+", " ", text).strip()

    parts = re.split(r"(?=Адреса\s+засновника:)", t)
    founders = []

    for part in parts:
        if not part.strip():
            continue
        m_name = re.search(r"^(.+?)\s+Адреса\s+засновника:", part, flags=re.I)
        name = m_name.group(1).strip(" ,;") if m_name else np.nan

        m_address = re.search(r"Адреса\s+засновника:\s*(.+?)(?=Розмір внеску|Частка|\Z)", part, flags=re.I)
        address = m_address.group(1).strip(" ,;") if m_address else np.nan

        m_amount = re.search(r"Розмір внеску.*?:\s*([\d\s,\.]+)\s*грн", part, flags=re.I)
        amount = np.nan
        if m_amount:
            try:
                amount = float(m_amount.group(1).replace(" ", "").replace(",", "."))
            except:
                pass

        m_share = re.search(r"Частка\s*\(%\):\s*([\d,\.]+)", part, flags=re.I)
        share = np.nan
        if m_share:
            try:
                share = float(m_share.group(1).replace(",", "."))
            except:
                pass

        founders.append((name, address, amount, share))

    while len(founders) < max_founders:
        founders.append((np.nan, np.nan, np.nan, np.nan))

    founders = founders[:max_founders]

    flat = [x for f in founders for x in f]
    return pd.Series(flat)


founder_cols = []
for i in range(1, 4):
    founder_cols.extend([
        f"Founder_name_{i}",
        f"Founder_address_{i}",
        f"Founder_amount_{i}",
        f"Founder_share_{i}"
    ])

df[founder_cols] = df["Перелік засновників/учасників юридичної особи"].apply(parse_multiple_founders)
df
```
