### Відповідальна персона (AUTHORISED PERSON)

```python
def parse_authorised_person(text):
    parts = [p.strip() for p in str(text).split(",")]
    records = []
    
    for p in parts:
        date_match = re.search(r"\d{2}\.\d{2}\.\d{4}", p)
        date = date_match.group(0) if date_match else "Невідомо"

        person_match = re.search(r"[А-ЯІЇЄҐ'\-]+\s+[А-ЯІЇЄҐ'\-]+\s+[А-ЯІЇЄҐ'\-]+", p)
        person = person_match.group(0).title() if person_match else "Невідомо"

        role = p
        if person_match:
            role = role.replace(person_match.group(0), "").strip()
        if date_match:
            role = role.replace(date_match.group(0), "").strip()
        role = role.split()[0] if role else ""

        records.append({"AUTH_PER": person, "PER_ENTRY_DATE": date, "PER_ROLE": role})
    return records
```

```python
rows = []
for _, row in df.iterrows():
    parsed = parse_authorised_person(row["AUTHORISED_PERSON"])
    for rec in parsed:
        new_row = row.drop(labels=["AUTHORISED_PERSON"]).to_dict()
        new_row.update(rec) 
        rows.append(new_row)

df = pd.DataFrame(rows)
```

## Активність  (парсинг Activity)

```python
def parse_activities(text):
    text = str(text).strip()
    records = []

    main_match = re.search(r"Основний:\s*(.+?)(?=Інші:|$)", text, re.DOTALL)
    other_match = re.search(r"Інші:\s*(.+)", text, re.DOTALL)

    if main_match:
        main_text = main_match.group(1).strip()
        m = re.match(r"(\d{2}\.\d{2})\s+(.+)", main_text)
        if m:
            records.append({
                "ACTIVITY_TYPE": "Основний",
                "KVED_CODE": m.group(1),
                "DESCRIPTION": m.group(2).strip()
            })

    if other_match:
        other_text = other_match.group(1).strip()
        parts = re.findall(r"(\d{2}\.\d{2})\s+([^0-9]+)(?=\d{2}\.\d{2}|$)", other_text)
        for code, desc in parts:
            records.append({
                "ACTIVITY_TYPE": "Інші",
                "KVED_CODE_OTH": code,
                "DESCRIPTION": desc.strip()
            })

    return records
```

```python
rows = []
for _, row in df.iterrows():
    parsed = parse_activities(row["ACTIVITES"])
    for rec in parsed:
        new_row = row.drop(labels=["ACTIVITES"]).to_dict()
        new_row.update(rec)
        rows.append(new_row)

df = pd.DataFrame(rows)

df
```
